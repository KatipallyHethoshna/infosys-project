```python
#step3
#In Data Cleaning we'll handle missing values, if any. we can check for missing data using isnull() and fill missing data using fillna().
#check for missing value
df.isnull()
```

```python
#if we find any missing values,we will fill the missing values in a column with the mean() or median()
#to fill missing values in a column with mean()
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
```

```python
#to fill the missing values in a column with median()
df['bmi'] = df['bmi'].fillna(df['bmi'].median())
```
